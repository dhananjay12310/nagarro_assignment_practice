# SNAKE AND LADDER GAME

```python
import random
```

```python
# board class to allow different snakes and ladders and size configuration
#default boardsize=10*10

class Board:
    def __init__(self, m=10,n=10):
        self.size = m*n
        self.snake_pos = dict()
        self.ladder_pos = dict()

    def add_snake(self, head,tail):
        self.snake_pos[head]=tail

    def add_ladder(self, start,end):
        self.snake_pos[start]=end
```

```python
class Dice:
    def roll():
        return random.randint(1,6)
```

```python
class Player:
    def __init__(self, name,player_pos=0):
        self.name = name
        self.player_pos=player_pos
        
    def update_player_pos(self, new_position):
        self.player_pos = new_position
```

```python
class Game:
    def __init__(self, board):
        self.board = board
        self.players_position = []
        
    def add_players(self, p):
        player_position = Player(p)
        self.players_position.append(player_position)
        
    def has_won(self, position):
        if position == self.board.size:
            return True
        return False
    
    def snake_or_ladder(self, new_position):
        if new_position in  self.board.snake_pos:
            return self.board.snake_pos[new_position]
        if new_position in self.board.ladder_pos:
            return self.board.ladder_list[new_position]
        
    def find_new_position(self, new_position):
        
        if self.snake_or_ladder(new_position):
            return self.snake_or_ladder(new_position)
        else:
            return new_position
        
    def start(self):
        still_playing = len(self.players_position)
        while(still_playing)>1:
            for player_position in self.players_position:
                dice_value = Dice.roll()
                current_position = player_position.player_pos
                new_position = current_position + dice_value

                if new_position < self.board.size:
                    new_position = self.find_new_position(new_position)
                    player_position.update_player_pos(new_position)
                    print("dice_value:",dice_value,'\n',player_position.name.name, ' : ', current_position, ' >>> ', new_position)

                if self.has_won(new_position):
                    print("dice_value:",dice_value,'\n',player_position.name.name, "wins!")
                    player_position.update_player_pos(new_position + 1)
                    still_playing -= 1
```
